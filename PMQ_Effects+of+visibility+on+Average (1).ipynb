```python
import math
import scipy.stats
import matplotlib.pyplot as plt
```

```python
mu, sigma,epsilon, sigma_epsilon, kappa = 2.1, 2, 0, 1, 2
```

```python
def E (mu,sigma,epsilon, sigma_epsilon, kappa, N):
    z=((((sigma_epsilon**2)/N+sigma**2)**(1/2))*(kappa-mu))/sigma**2
    probability = 1-scipy.stats.norm(0, 1).cdf(z)
    return probability
```

```python
E_list=[]
N=111
for n in range (1,N):
    E_list.append(E(mu,sigma,epsilon, sigma_epsilon, kappa, n))
```

```python
plt.plot(range (1,N), E_list)
plt.xlabel('Number of visibility') 
plt.ylabel('Probability of becoming a minister') 
plt.show()
```
