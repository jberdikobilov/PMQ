```python
import math
import scipy.stats
import matplotlib.pyplot as plt
```

```python
mu, mu_i, sigma, epsilon, sigma_epsilon, kappa = 1, 1.9, 3, 0, 5, 2
```

```python
def E (mu, mu_i, sigma, epsilon, sigma_epsilon, kappa, N):
    z=(sigma_epsilon*(kappa-mu))/(math.sqrt(N)*sigma**2)+(math.sqrt(N)*(kappa-mu_i))/sigma_epsilon
    probability = 1-scipy.stats.norm(0, 1).cdf(z)
    return probability
```

```python
E_list=[]
N=101
for n in range (1,N): 
    E_list.append(E(mu,mu_i, sigma,epsilon, sigma_epsilon, kappa, n))
```

```python
plt.plot(range (1,N), E_list)
plt.xlabel('Number of visibility') 
plt.ylabel('Probability of becoming a minister') 
plt.show()
```
